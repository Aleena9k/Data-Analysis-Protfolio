### Filling the customer ID with Guest

### Changing Data Types

```python
df = df[df['InvoiceDate'] != 'InvoiceDate']
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce', infer_datetime_format=True)
```

```python
df.info()
```

```python
df
```

```python
df['Quantity'] = pd.to_numeric(df['Quantity'])
```

### Duplicate records

### Removing Invalid transactions

**UNIT PRICE EQUALS 0**

```python
df = df[df['UnitPrice'] != 0]
```

```python
df
```

```python
df[df['UnitPrice'] < 0]
```

**CANCELLATION DATA**

```python
cancellation_data = df[df['InvoiceNo'].astype(str).str.startswith('C')]
cancellation_data
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
df
```

```python
df[df['InvoiceNo'].astype(str).str.startswith('C')]
```

**ADJUST BAD DEBT**

```python
df = df[~df['Description'].str.contains('Adjust bad debt', case=False, na=False)]
```

```python
df
```

### Adding new columns

**TOTAL PRICE COLUMN**

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

**SEPARATING YEARS AND MONTHS**

```python
df['InvoiceYearMonth'] = df['InvoiceDate'].dt.to_period('M')
df['InvoiceYear'] = df['InvoiceDate'].dt.year
df['InvoiceMonth'] = df['InvoiceDate'].dt.month
```
